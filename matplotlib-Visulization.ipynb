# Visulization

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
years = [1960, 1970, 1980, 1990, 2000, 2010, 2016]
iran_pop = [21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28]
```

# Matplotlib

#### Line Plot

```python
plt.plot(years, iran_pop)
plt.show()
```

#### Scatter Plot

```python
plt.scatter(years, iran_pop)
plt.show()

```

#### Histogram Plot

```python
#Histogram Help to Get Idea About Distribution

city_name = ['Tehran', 'Mashhad', 'Isfahan', 'Karaj', 'Tabriz', 'Shiraz']
city_pop = [7153309, 2307177, 1547164, 1448075, 1424641, 1249942]
```

```python
plt.hist(city_pop)
plt.show()
```

###### The "square root rule" is a commonly-used rule of thumb for choosing number of bins

The bins parameter tells you the number of bins that your data will be divided into. set bins=1 to understand it

```python
plt.hist(city_pop,bins=1)
plt.show()
```

###### قراردادن
bins
به مقدار حدودی جذر داده ها مقدار مناسبی هست و با شکل بهتری توزیع اطلاعات رو نشون میده
مثلا الان جذر 6 رو به سمت پایین رد کنی میشه دو
پس:
    bins=2




```python
plt.hist(city_pop,bins=2)
plt.show()
```

###### محاسبه مقدار مناسبbins به طور اتوماتیم

```python
plt.hist(city_pop,bins='auto')
plt.show()
```

#### Pie Chart

```python
plt.pie(city_pop,labels=['Tehran', 'Mashhad', 'Isfahan', 'Karaj', 'Tabriz', 'Shiraz'] )
plt.show()
```

#### Customization plots

```python
#Customization plots

plt.figure(figsize=(9, 7), dpi=90)
plt.plot(years, iran_pop, color='green')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Iran\'s Population')
plt.yticks([21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28],['21M','28M','38M','56M',
'66M','74M','80M'])
plt.xticks([1960, 1970, 1980, 1990, 2000, 2010, 2016] )
plt.show()
```

```python
# Customize Scatter plot
# در محور x
# به جای لیست شهر ها از 0 تا 5 یعنی 6 عدد گذشتیم به کمک 
#np.arange(6)

plt.scatter(np.arange(6),city_pop)
plt.show()
```

```python
# Customize Scatter plot

popsize = np.array(iran_pop)*3

plt.scatter(years, iran_pop, s=popsize)

# s=popsize
#

plt.xlabel('Year')
plt.ylabel('Population')

plt.title('Iran\'s Population')

plt.yticks([21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28],['21M','28M','38M','56M',
'66M','74M','80M'])
plt.xticks([1960, 1970, 1980, 1990, 2000, 2010, 2016] )

plt.show()
```

```python
pop_size = np.array([7153309, 2307177, 1547164, 1448075, 1424641, 1249942])/30000
Colors = ['violet', 'green', 'orange', 'tomato', 'blue', 'pink']

plt.scatter(np.arange(6),city_pop, s=pop_size,c=Colors, alpha=1)

plt.xticks([0, 1, 2, 3, 4, 5],['Tehran', 'Mashhad', 'Isfahan', 'Karaj', 'Tabriz', 'Shiraz'],rotation=60,fontsize='large')
plt.yticks([1000000, 2000000, 7000000],['1M','2M','7M'])

plt.title('Iran\'s most populated city')

plt.text(0,7153309,'iran Capital',fontsize=15)
plt.ylim([0,8000000])
plt.show()
```
