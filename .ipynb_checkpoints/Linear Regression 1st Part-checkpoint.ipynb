```python
import numpy as np
```

```python
#x,y لیست هستند

x=[50,53,55,60,64,70,84,88,90,94,100,110,115,120,145,150,200]
y=[80,75,78,82,90,85,80,90,80,100,80,120,150,180,200,150,320]

plt.figure()
```

```python
plt.plot(x,y,'rx')
```

```python
#ایجاد هزار عدد در یک نامپی ارایه بین 0-250 با فاصله یکسان

line_x=np.linspace(0,250,1000)
```

```python
plt.plot(line_x,line_x)
```

```python
# سوال:جرا خط رو با تعداد زیادی نقطه رسم میکنیم؟؟؟؟؟؟

line_x=np.linspace(0,1000,10000)
plt.plot(x,y,'rx')
plt.plot(line_x,line_x)
```

```python
#تعریف محدوده برای اینکه دور نشیم از محدوده x,y تعریف شده
plt.xlim(0,250)
plt.ylim(0,350)

line_x=np.linspace(0,1000,10000)
plt.plot(x,y,'rx')
plt.plot(line_x,line_x)
```

```python
plt.xlim(0,250)
plt.ylim(0,350)

line_x=np.linspace(0,1000,10000)
plt.plot(x,y,'rx')

#ضریب نیم و تغییر شیب خط

plt.plot(line_x,0.5*line_x)
plt.plot(x,y,'rx')
```

```python
#برچسب سطر ستون برای نمودار

plt.xlabel('size($m^2$)')
plt.ylabel( ' Prices (Million Toman) ')

plt.plot(line_x,0.5*line_x)
plt.plot(x,y,'rx')
```

```python
def f(x):
    theta0=-20
    theta1=2
    return theta1*x+theta0  
    
```

```python
plt.plot(line_x,f(line_x),'g')
```
