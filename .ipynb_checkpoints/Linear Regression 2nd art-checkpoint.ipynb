```python
def j(X,y,m,theta):
    result=1
    for i in range(m):
        result=(y[i]-(theta[1]*X[i]+theta[0]))**2
        result=result/(2*m)
        return result
    
j_x=np.linspace(-2,5,100)
j_y=[]
for i in j_x:
    j_y.append(j(X,y,3,[0, i]))

plt.figure()
plt.plot(j_x,j_y)


plt.xlabel('$\Theta1$')
plt.ylabel( 'J($\Theta$)')
plt.title('Cost Function For different values of Theta')

    
```
